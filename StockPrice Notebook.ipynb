```python
import sys
import os
from urllib.request import Request
from urllib.request import urlopen
from urllib.error import URLError
import json
import datetime
from dateutil.relativedelta import relativedelta
```

```python
def  openAndClosePrices(firstDate, lastDate, weekDay):
    #constants of the function
    url = "https://jsonmock.hackerrank.com/api/stocks/"
    oneMonth = relativedelta(months=1)
    dateFormat = "%d-%B-%Y"
    weekDayList = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday']
    
    #creating datetime using the service format
    dateTimeMin = datetime.datetime.strptime(firstDate, dateFormat)
    dateTimeMax =  datetime.datetime.strptime(lastDate, dateFormat)
    
    try:
        while dateTimeMin <= dateTimeMax:
            pageIndex = 1
            totalPages = 1
            while pageIndex <= totalPages:
                #getting the named month to pass on the urlFilter
                namedMonth = dateTimeMin.strftime('%B')
                urlFilter = url + "search?date="+ namedMonth + "-" + str(dateTimeMin.year) + "&page=" + str(pageIndex)
                urlResponse = urlopen(urlFilter)
                
                if urlResponse.getcode() == 200:
                    
                    jsonResponse = json.loads(urlResponse.read())
                    
                    #total_pages for pagination
                    totalPages = jsonResponse["total_pages"]
                    dataCollection = jsonResponse["data"]
                    
                    #getting all the stock data that is between the first and last date and printing
                    for data in dataCollection:
                        dateData = datetime.datetime.strptime(data.get("date"), dateFormat)
                        weekDayPosition = dateData.weekday()
                        
                        if dateTimeMin <= dateData <= dateTimeMax:
                            #veryfing the weekday
                            if  weekDayList[weekDayPosition] == weekDay:
                                joinSequence = (data.get("date"), str(data.get("open")), str(data.get("close")))
                                print(" ".join(joinSequence))
                pageIndex += 1;
            
            #adds one more month and adjust to day 1
            dateTimeMin = (dateTimeMin + oneMonth).replace(day=1)
                              
    except URLError as e:
        print(str(e))
    except Exception as e:
        print("Message: Internal Error")
    
 
try:
    _firstDate = '26-March-2000'
except:
    _firstDate = None


try:
    _lastDate = '15-August-2000'
except:
    _lastDate = None


try:
    _weekDay = 'Wednesday'
except:
    _weekDay = None

openAndClosePrices(_firstDate, _lastDate, _weekDay)
```
